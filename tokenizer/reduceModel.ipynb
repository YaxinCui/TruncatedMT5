```python
from bert4keras.tokenizers import SpTokenizer
import json
```

```python
from bert4keras.models import build_transformer_model
```

```python
# 模型路径
config_path = '../mT5Large/keras_config.json'
checkpoint_path = '../large'
spm_path = './sentencepiece_cn.model'
keep_tokens_path = './sentencepiece_cn_keep_tokens.json'

# 加载分词器
tokenizer = SpTokenizer(spm_path, token_start=None, token_end='&lt;/s&gt;')
keep_tokens = json.load(open(keep_tokens_path))

# 加载模型
t5 = build_transformer_model(
    config_path=config_path,
    checkpoint_path=checkpoint_path,
    keep_tokens=keep_tokens,
    model='t5.1.1',
    return_keras_model=False,
    name='T5',
)

encoder = t5.encoder
decoder = t5.decoder
model = t5.model
```
